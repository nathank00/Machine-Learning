```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
def make_data():
    apple = pd.read_csv('AAPL.csv')
    intel = pd.read_csv('INTC.csv')
    cola = pd.read_csv('KO.csv')
    micro = pd.read_csv('MSFT.csv')
    tesla = pd.read_csv('TSLA.csv')

    apple['Date'] = pd.to_datetime(apple['Date'])
    intel['Date'] = pd.to_datetime(intel['Date'])
    cola['Date'] = pd.to_datetime(cola['Date'])
    micro['Date'] = pd.to_datetime(micro['Date'])
    tesla['Date'] = pd.to_datetime(tesla['Date'])

    tesla_cl = tesla['Adj Close'].pct_change()
    apple_cl = apple['Adj Close'].pct_change()
    cola_cl = cola['Adj Close'].pct_change()
    micro_cl = micro['Adj Close'].pct_change()
    intel_cl = intel['Adj Close'].pct_change()
    date = apple['Date']

    #print(tesla_cl.iloc[-1], apple_cl.iloc[-1], cola_cl.iloc[-1], micro_cl.iloc[-1], intel_cl.iloc[-1])
    return_mat = pd.DataFrame(data=[tesla_cl,apple_cl,cola_cl,micro_cl,intel_cl])
    return_mat = return_mat.T
    return_mat.columns = ['Tesla','Apple','Cola','Micro','Intel']

    return return_mat, date

data, index = make_data()
data.index = index
data
```

```python
data = data.dropna()
data
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
num_components = 5

# Create PCA object
pca = PCA(n_components=num_components)

# Fit and transform the data
principal_components = pca.fit_transform(data)
```

```python
pc_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(num_components)])

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f'Explained Variance Ratio: {explained_variance_ratio}')

# Display the DataFrame with principal components
print(pc_df)
```

```python
loadings_df = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(num_components)], index=data.columns)
print('Loadings of Original Features on Principal Components:')
print(loadings_df)
```
